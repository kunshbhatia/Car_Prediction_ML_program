```python
import pandas as pd
```

```python
df = pd.read_csv("Vehicle Price.csv")
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop(["name","description",'exterior_color',"interior_color"],axis=1,inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['price'] = df['price'].fillna(df['price'].mode()[0])
```

```python
df['fuel'] = df['fuel'].fillna('Electric') # As all cars having nan are electric , you can check also
```

```python
df['mileage'] = df['mileage'].fillna(df['mileage'].median())
```

```python
# Checked the correct data from internet
df.loc[725,"transmission"] = "1-Speed Automatic"
df.loc[940,"transmission"] = '10-Speed Automatic with Overdrive'
df['doors'] = df['doors'].fillna(4)
```

```python
df['body'] = df['body'].fillna(df['body'].mode()[0])
```

```python
df.isnull().sum() # All Set
```

```python
df.reset_index(drop=True,inplace=True)
df
```

```python
df.to_csv("Vehicle_final.csv") # Saving The Fresh Data
```
